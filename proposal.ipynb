## Preliminary exploratory data analysis

```r
library(tidyverse)
library(repr)
library(tidymodels)
library(rvest)
install.packages("GGally")
library("GGally")
options(repr.matrix.max.rows = 6)
```

Below, we're looking at the correlation between the variables to try to pick the appropriate predictors from the available variables. 

```r
shoppers_data_full <- read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00468/online_shoppers_intention.csv")
shoppers_data_numeric <- shoppers_data_full %>% 
    mutate(Revenue = ifelse(Revenue==TRUE, 1, 0))

#  the visualization of correlation coefficients between each numerical variable in our data and our target, Revenue
shoppers_cor <- ggcorr(shoppers_data_numeric, hjust = 1, size = 4, label = TRUE)
shoppers_cor
```

```r
# we choose the variables with top 5 correlation coefficient as our predictors
shoppers_data <- shoppers_data_full %>% 
    select(ProductRelated,ProductRelated_Duration,BounceRates,ExitRates,PageValues,Revenue) %>% 
    mutate(Revenue = as_factor(Revenue))
shoppers_data
```

The dataset is already in a tidy format where each row is a single observation (data of a single sample shopper visiting an online shopping page).

Each column is a single varible:
- ProductRelated: Number of pages of related products visited by a sample shopper
- ProductRelated_Duration: Time spent on pages of related products
- BounceRates: the average percentage of people who enter from that page and leave without taking other actions
- ExitRates: the average percentage that the page in the last page visited before the shopper exits.
- PageValues:
- Revenue: categorial variable indicates wheter a visitor end up buying the products (FALSE: did not buy, TRUE: did buy)

Finally, each value is in a single cell.

```r
summary(shoppers_data)
print('The number of cells with missing value:')
      sum(is.na(shoppers_data))
print('Total number of observations:')
      nrow(shoppers_data)
```

```r
set.seed(1)

shoppers_split <- initial_split(shoppers_data, prop = 0.75, strata = Revenue)
shoppers_train <- training(shoppers_split)
shoppers_test <- testing(shoppers_split)
```

Shown below are the example visualization, which graphs the distributions of each of the predictors.

```r
options(repr.plot.height = 6, repr.plot.width = 15)
page_values_hist <- ggplot(shoppers_train, aes(x = PageValues, fill = Revenue)) +
    geom_histogram(position = "identity", binwidth = 30) +
    facet_grid(. ~ Revenue) +
    labs(x = "Page Values", y = "Visitors", fill = "Does the visitor buy the product") +
    theme(text = element_text(size = 15))
page_values_hist
```

# Methods
K-nearest neighbours will be the main model used for this project. The correlation coefficient of all numeric attributes were calculated and visualized in a correlation matrix. The top five correlations were chosen as our predictors. The model will be trained by conducting 5 individual k-nearest neighbours on each of the chosen attribute. Once the model has been trained using the training set (75% of the original data set), the overall training data set (25% of the original data set) will be divided into two subsets – training set and validation set. Then we will calculate the accuracy estimate and plot them against the number of neighbours (K). Finally, we will verify whether our model and answer our question.

# Expected outcomes and significance

The five attributes chosen, although are correlated to some degree, are not very strong. This makes it difficult to predict whether or not the model will be successful. However, we expect that the model will predict results better than random chance. 

The findings of the project will be significant in understanding consumer purchasing behaviour online, which can be used by businesses to tailor the customer’s online experience in aims of maximizing revenue and profit.
